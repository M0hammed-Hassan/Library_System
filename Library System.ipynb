```python
class Library:
    requested_books=[] #Class attribute.
    def __init__(self,availablebooks): #Instance method include the available books in the library.
        self.availablebooks=availablebooks #I make this attribute private.
    def create_account(self): #This instance method for making account for the users.
        print("================================")
        print('Create account to continue')
        self.name=input('Enter your name:')
        while True:
            try:
                self.age=int(input('Enter your age:'))
                break
            except:
                print('Age must be a number, try again')
        
        self.gender=input('Enter your gender:')
    def account_details(self): #This instance method for user information.
        print(f"Account details:\nYour name: {self.name}\nYour age: {self.age}\nYour gender: {self.gender}")
        print('Borrowed books:')
        if len(Library.requested_books)==0:
            print('   You don\'t have books')
        else:
            for i in Library.requested_books:
                print('   ',i)
    def interface(self): #This instance method for our interface after creating account.
        print('        LIBRARY MENU        ')
        print('1.Account details')
        print('2.Display all available books')
        print('3.Request a book')
        print('4.Return a book')
        print('5.Add a book')
        print('6.Exit')
    def request_book(self): #This instance method for borrow book
        book_name=input('Enter the name of the book you wanna borrow it:')
        if book_name in self.availablebooks: #We will check this book available, if available we will remove it.
            print('The book you requested has now been borrowed')
            self.availablebooks.remove(book_name)
            Library.requested_books.append(book_name)
        else:
            print('This book not available now')
    def return_book(self): #This instance method for returning borrowed books
        book_name=input('Enter the name of the book you wanna return it:')
        if book_name in self.requested_books: #We will see this is our book or not.
            self.availablebooks.append(book_name)
            self.requested_books.remove(book_name)
            print('Thanks for returning your borrowed book')
        else:
            print('Incorrect book, review your borrowed books')
    def display_books(self): #This instance method for dispaly our available books.
        print('Avaliable books are:')
        for idx,book in enumerate(self.availablebooks,1):
            print(f"   {idx}-{book}")
    def add_book(self,book): #This instance method for adding a book in the library repository.
        self.availablebooks.append(book)
        print('Thanks for adding a book')
    def remove_storage(): #For clearing requested_books list before joining another user.
        Library.requested_books.clear()
Books=["The Last Battle","The Screwtape letters","The Great Divorce"]
```

```python
class Student(Library): #Class student will inherit class library.
    def __init__(self,availablebooks):
        super().__init__(availablebooks) #For reusing __init__ method.
    def student_attributes(self):#Instance method include attributes for students.
        while True:
            try:
                self.ID=int(input('Enter your id:'))
                break
            except:
                print('Incorrect id, try again')
        self.faculty=input('Enter your faculty')
        self.department=input('Enter your department:')
    def studen_information(self): #This instance method return student information.
        return f"Your id:{self.ID}\nYour faculty:{self.faculty}\nYour department:{self.department}"
    def student_choice(self): #This instance method help user to choice a new operation after the previous one.
        self.interface()
        while True:
            try:
                self.choice=int(input('Enter your choice:'))
                break
            except:
                print('Incorrect choice, try again')
    def main(self): #This instance method for run student class.
        self.create_account()
        self.student_attributes()
        print("Account created successfully")
        print("================================")
        while True:
            self.student_choice()
            if self.choice==1:
                self.account_details()
                print(self.studen_information())
                print("================================")
            elif self.choice==2:
                self.display_books()
                print("================================")
            elif self.choice==3:
                self.request_book()
                print("================================")
            elif self.choice==4:
                self.return_book()
                print("================================")
            elif self.choice==5:
                self.add_book(input('Enter a book:'))
                print("================================")
            elif self.choice==6:
                print(f'See you later {self.name}')
                print("================================")
                return
student=Student(Books)
```

```python
class Author(Library): #Class author will inherit class library.
    def __init__(self,availablebooks):
        super().__init__(availablebooks) #For reusing __init__ method.
    def author_attributes(self): #This instance method author attributes.
        self.specialization=input('Enter your specialization:')
        self.book=input('Enter your book:')
    def author_info(self): #This instance method return author information.
        return f"Your specialization:{self.specialization}\nYour book:{self.book}"
    def author_choice(self): #This instance method help user to choice a new operation after the previous one.
        self.interface()
        while True:
            try:
                self.choice=int(input('Enter your choice:'))
                break
            except:
                print('Incorrect choice, try again')
    def main(self): #This instance method for run author class.
        self.create_account()
        self.author_attributes()
        print("Account created successfully")
        print("================================")
        while True:
            self.author_choice()
            if self.choice==1:
                self.account_details()
                print(self.author_info())
                print("================================")
            elif self.choice==2:
                self.display_books()
                print("================================")
            elif self.choice==3:
                self.request_book()
                print("================================")
            elif self.choice==4:
                self.return_book()
                print("================================")
            elif self.choice==5:
                self.add_book(self.book)
                print('Your book has been added successfully')
                print("================================")
            elif self.choice==6:
                print(f'See you later {self.name}')
                print("================================")
                return
author=Author(Books)
```

```python
def main():
    print('          Welcome to our library          ')
    while True:
        try:
            choice=int(input('Press 1 if you\'re a student or press 2 if you\'re author:'))
            break
        except:
            print('Incorrect choice, try again')
    if choice==1:
        Library.remove_storage()
        student.main()
    elif choice==2:
        Library.remove_storage()
        author.main()
main()
```
